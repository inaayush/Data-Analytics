```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from mpl_toolkits import mplot3d
```

## Multiple Regression

```python
data = pd.read_csv("2_vaiable_mr.csv")
y = np.array(data["Y"])#enter y value here
X1 = np.array(data["X1"])#enter X1 value here
X2 = np.array(data["X2"])#enter X2 value here
```

```python
def Multiple_regression_for_two_variables(y,X1,X2):
    data = pd.DataFrame()
    data["y"] = y
    data["X1"] = X1
    data["X2"] = X2
    data["X1^2"] = X1**2
    data["X2^2"] = X2**2
    data["X1y"] = X1*y
    data["X2y"] = X2*y
    data["X1X2"] = X1*X2
    sX1p2 = np.sum(data["X1^2"])-np.sum(data["X1"])**2/data.shape[0]
    sX2p2 = np.sum(data["X2^2"])-np.sum(data["X2"])**2/data.shape[0]

    sX1y = np.sum(data["X1y"]) - (np.sum(data["X1"])*np.sum(data["y"])/data.shape[0])
    sX2y = np.sum(data["X2y"]) - (np.sum(data["X2"]*np.sum(data["y"]))/data.shape[0])

    sX1X2 = np.sum(data["X1X2"])-np.sum(data["X1"])*np.sum(data["X2"])/data.shape[0]

    print("ΣX1^2:",sX1p2,"    ΣX2^2:",sX2p2,"         ΣX1y:",sX1y,"        ΣX2y:",sX2y,"       ΣX1X2:",sX1X2)
    #values of B1 and B2
    B1 = (sX2p2*sX1y - sX1X2*sX2y)/(sX1p2*sX2p2-sX1X2**2)
    B2 = (sX1p2*sX2y - sX1X2*sX1y)/(sX1p2*sX2p2-sX1X2**2)
    a = np.mean(data["y"])-B1*np.mean(data["X1"])-B2*np.mean(data["X2"])
    print("\nβ1:",B1,"   β2:",B2,"       a:",a)
    #equation
    print("\nEquation: y=",a,"+",B1,"x1",B2,"x2","+ e")
    #predictive value
    yp = B1*(X1) + B2*(X2) + a
    print("\nyp:",yp)
    #least Square value 
    least_square = np.sum((yp-np.mean(y))**2)/np.sum((y-np.mean(y))**2)
    print("\nleast square:",least_square)
    xyplot = np.array([X1,X2])
    zxplot = np.array([X1,yp])
    yzplot = np.array([X2,yp])
    fig = plt.figure(figsize=[7,6])
    ax = fig.add_subplot(111, projection='3d')
#     ax.scatter3D(data["X1"],data["X2"], data["y"],c=y, cmap='Greens')
    sb.scatterplot(x=X1,y=yp)
    sb.scatterplot(x=X2,y=yp)
    sb.scatterplot(x=X2,y=X1,color="blue")
    plt.xlabel("X")
    plt.ylabel("Y")
    ax.plot_surface(xyplot, yzplot,zxplot,alpha=0.2,)
    data.to_csv("output_filename.csv", index=False, encoding='utf8')
```

```python
Multiple_regression_for_two_variables(y,X1,X2)
```
